#table 01 Analysis



```python
# Top growing sectors
merchandise_sectors = df1.iloc[0:17].copy()

top_growth_sectors = merchandise_sectors.sort_values(by='% Growth (23-24)', ascending=False).head(5)
print("Top 5 Growing Sectors (Jan-Dec):")
print(top_growth_sectors[['Exports of Goods & Services', '% Growth (23-24)']])

top_growth_dec = merchandise_sectors.sort_values(by='% Growth (23-24) Dec', ascending=False).head(5)
print("\nTop 5 Growing Sectors (Dec):")
print(top_growth_dec[['Exports of Goods & Services', '% Growth (23-24) Dec']])

plt.figure(figsize=(12, 8))
sns.barplot(x='% Growth (23-24)', y='Exports of Goods & Services', data=merchandise_sectors.sort_values('% Growth (23-24)', ascending=False))
plt.title('Sector-wise % Growth (Jan-Dec 2023-2024)')
plt.show()
```

```python
#Analysis for Table 1 - Contribution to Total Exports

merchandise_2024 = df1[df1['Exports of Goods & Services'] == 'Total Merchandise Exports']['Jan-Dec 2024'].values[0]
services_2024 = df1[df1['Exports of Goods & Services'] == 'Total Services Exports']['Jan-Dec 2024'].values[0]

print(f"Merchandise Exports 2024: {merchandise_2024} ({merchandise_2024 / (merchandise_2024 + services_2024) * 100:.2f}%)")
print(f"Services Exports 2024: {services_2024} ({services_2024 / (merchandise_2024 + services_2024) * 100:.2f}%)")

top_sectors_2024 = df1.iloc[0:5][['Exports of Goods & Services', 'Jan-Dec 2024']]
others_2024 = merchandise_2024 - top_sectors_2024['Jan-Dec 2024'].sum()
top_sectors_2024 = pd.concat([top_sectors_2024, pd.DataFrame({'Exports of Goods & Services': ['Others'], 'Jan-Dec 2024': [others_2024]})])

plt.figure(figsize=(8, 8))
plt.pie(top_sectors_2024['Jan-Dec 2024'], labels=top_sectors_2024['Exports of Goods & Services'], autopct='%1.1f%%')
plt.title('Top Sectors Contribution to Merchandise Exports 2024')
plt.show()
```

#Table 02 Analize

```python
#Analysis for Table 2 - Top Destinations

destinations = df2.iloc[0:15].copy()

top_dest_2024 = destinations.sort_values(by='Jan-Dec 2024', ascending=False).head(10)
print("Top 10 Destinations by Value 2024:")
print(top_dest_2024[['Destination', 'Jan-Dec 2024']])

plt.figure(figsize=(12, 8))
sns.barplot(x='Jan-Dec 2024', y='Destination', data=top_dest_2024)
plt.title('Top 10 Export Destinations 2024')
plt.show()

# Growth analysis
top_growth_dest = destinations.sort_values(by='% Growth (23-24)', ascending=False).head(5)
print("\nTop 5 Growing Destinations:")
print(top_growth_dest[['Destination', '% Growth (23-24)']])
```

```python
#Analysis for Table 3 - Region-wise

regions = df3.iloc[0:8].copy()

top_growth_regions = regions.sort_values(by='% Growth (23-24)', ascending=False).head(5)
print("Top Growing Regions (Jan-Dec):")
print(top_growth_regions[['Region', '% Growth (23-24)']])

plt.figure(figsize=(8, 8))
plt.pie(regions['Jan-Dec 2024'], labels=regions['Region'], autopct='%1.1f%%')
plt.title('Region-wise Export Distribution 2024')
plt.show()

top_growth_dec_regions = regions.sort_values(by='% Growth (23-24) Dec', ascending=False).head(5)
print("\nTop Growing Regions (Dec):")
print(top_growth_dec_regions[['Region', '% Growth (23-24) Dec']])
```

```python
#Comparative Analysis
us_table2 = df2[df2['Destination'] == 'United States']['Jan-Dec 2024'].values[0]
us_table3 = df3[df3['Region'] == 'United States']['Jan-Dec 2024'].values[0]
print(f"US Export 2024 consistent: {us_table2 == us_table3}")

# Total consistency
total_merch_1 = df1[df1['Exports of Goods & Services'] == 'Total Merchandise Exports']['Jan-Dec 2024'].values[0]
total_merch_2 = df2[df2['Destination'] == 'Total']['Jan-Dec 2024'].values[0]
total_merch_3 = df3[df3['Region'] == 'Total Merchandise Exports']['Jan-Dec 2024'].values[0]
print(f"Totals consistent: {total_merch_1 == total_merch_2 == total_merch_3}")
```

```python
#Advanced Analysis - Correlation between full year and Dec growth for sectors

merchandise_sectors['Sector'] = merchandise_sectors['Exports of Goods & Services']
corr = merchandise_sectors[['% Growth (23-24)', '% Growth (23-24) Dec']].corr()
print("Correlation between full year and Dec growth:")
print(corr)

# Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='% Growth (23-24)', y='% Growth (23-24) Dec', data=merchandise_sectors)
plt.title('Full Year vs Dec Growth Correlation')
plt.show()
```
